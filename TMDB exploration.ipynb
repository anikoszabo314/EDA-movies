```python
# This function creates a new DataFrame showing the number and percentage of null values in each column

def null_vals(dataframe):
    '''function to show both number of nulls and the percentage of nulls in the whole column'''
    null_vals = dataframe.isnull().sum() ## How many nulls in each column
    total_cnt = len(dataframe) ## Total entries in the dataframe
    null_vals = pd.DataFrame(null_vals,columns=['null']) ## Put the number of nulls in a single dataframe
    null_vals['percent'] = round((null_vals['null']/total_cnt)*100,3) ## Round how many nulls are there, as %, of the df
    
    return null_vals.sort_values('percent', ascending=False)
```

```python
# Apply the function to our dataset

null_vals(df)
```

### 2.1 Null values

There are two columns with a <b>high percentage</b> of null values:
<ul>
    <li>homepage</li>
    <li>tagline</li>
</ul>

We need to <b><u>fill</u></b> these missing values with something (I chose an X).

There are three additional columns with a <b>very low percentage</b> of null values:
<ul>
    <li>overview</li>
    <li>runtime</li>
    <li>release_date</li>
</ul>

We can <b><u>drop</u></b> these three columns because they represent such a small fraction of the dataset.

```python
# Drop the rows containing null values in the three columns mentioned above

df.dropna(
    subset = ['overview', 'runtime', 'release_date'],   # subset tells dropna which columns to consider
    inplace = True # makes change permanent
)
```

```python
# Make sure that we no longer have nulls in those three columns

null_vals(df)
```

```python
# Fill the missing homepage values with X

# Step 1: turn into category data type
df['homepage'] = df.homepage.astype('category')

# Step 2: add X category
df['homepage'] = df['homepage'].cat.add_categories('X')

# Step 3: fill the missing cells with an X
df['homepage'] = df['homepage'].fillna(value = 'X', method = None) 
```

```python
# Check that we no longer have nulls in the homepage column

null_vals(df)
```

```python
#  Fill the missing tagline values with X

# Step 1: turn into category data type
df['tagline'] = df.tagline.astype('category')

# Step 2: add X category
df['tagline'] = df['tagline'].cat.add_categories('X')

# Step 3: fill the missing cells with an X
df['tagline'] = df['tagline'].fillna(value = 'X', method = None)
```

```python
# Check that we no longer have nulls in the dataset

null_vals(df)
```

### 3.1 Extracting the year from the release_date column

```python
# Defining a function that extracts the year from a date given in the YYYY-MM-DD format

def extract_year(x):
    x = str(x)    # Converts the date to a string
    return int(x[0:4])    # Gets the first 4 characters (the year) and casts it into an integer
```

```python
# Apply the function to the dataset

df['release_year'] = df['release_date'].apply(extract_year)
```

Now we have a new column in the DataFrame called release_year.

### 3.2 The number of films released per year

```python
# The number of films per year, sorted from highest to lowest

movies_per_year = df.groupby('release_year')[['id']]\
                    .count()\
                    .sort_values(by=['id'],ascending=False)\
                    .copy()
movies_per_year.head()
```

```python
# Plot the number of films released per year

plt.figure(figsize = (10,5))    # Set the size of the output

sns.lineplot(
    x = movies_per_year.index, 
    y = movies_per_year['id'])

plt.title("The number of films released per year")    # The name of the plot

# Set axis labels
plt.xlabel("Release year")
plt.ylabel("Number of films")

plt.savefig('plot1.png', dpi = 300, bbox_inches = 'tight')    # save the output as a .png file

plt.show()
```

```python
## Function to extract genres for each datapoint
import ast ## This module can be used to evaluate literals, eg: transform string-lists back into lists
def extract_genres(x): 
    x = ast.literal_eval(x) ## Transform '['name', 'id']' back into ['name', 'id']
    Genres = [] ## Empty list to store the genres
    
    ''' Iterate through each dictionary '''
    for item in x: ## iterate for each dictionary in our list
        Genres.append(item['name']) ## Grab the 'name' key for each dictionary
        
    ''' Return the Genres'''
    return Genres

## Apply the function to genres
df['extracted_genres'] = df['genres'].apply(extract_genres)
```

```python
# Function to count the number of elements in a list

def count_list_length(x):
    return len(x)
```

```python
# Apply the function to the new number_of_genres column

df['number_of_genres'] = df['extracted_genres'].apply(count_list_length)
```
